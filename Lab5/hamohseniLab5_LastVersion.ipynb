```python
class BootCampTask(QCAlgorithm):

    def Initialize(self):
        self.SetStartDate(2020, 11, 1)  # Set Start Date
        self.SetEndDate(2021, 5, 1) #Set End Date
        self.SetCash(1000000)  # Set Strategy Cash
        
        
        self.iwm = self.AddEquity("IWM", Resolution.Minute)
        self.iwm.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.aapl = self.AddEquity("AAPL", Resolution.Minute)
        self.aapl.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.fubo = self.AddEquity("FUBO", Resolution.Minute)
        self.fubo.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.ge = self.AddEquity("GE", Resolution.Minute)
        self.ge.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.pltr = self.AddEquity("PLTR", Resolution.Minute)
        self.pltr.SetDataNormalizationMode(DataNormalizationMode.Raw)


    def OnData(self, data):

        if not self.Portfolio.Invested:
            self.SetHoldings("IWM", 0.2)
            self.SetHoldings("AAPL", 0.2)
            self.SetHoldings("FUBO", 0.2)
            self.SetHoldings("GE", 0.2)
            self.SetHoldings("PLTR", 0.2)
            
            
            self.Log("Active Pos iShare " + str(self.Portfolio["IWM"].Quantity));
            self.Log("Active Pos Apple " + str(self.Portfolio["AAPL"].Quantity));
            self.Log("Active Pos FuboTV " + str(self.Portfolio["FUBO"].Quantity));
            self.Log("Active Pos General Electric " + str(self.Portfolio["GE"].Quantity));
            self.Log("Active Pos Palantir Technologies " + str(self.Portfolio["PLTR"].Quantity));
```
